```python
#using log transformer method to tranform thhe data
data['Standard Error']=np.log(data['Standard Error'])
data['Family']=np.log(data['Family'])
data['Health (Life Expectancy)']=np.log(data['Health (Life Expectancy)'])
data['Trust (Government Corruption)']=np.log(data['Trust (Government Corruption)'])
data['Generosity']=np.log(data['Generosity'])
```

```python
data.head()
```

```python
#checking data types of each column
data.info()
```

# we can see that in column family,Health (Life Expectancy),Trust (Government Corruption and  Generosity its showing less values, so there might be infinty number present in it

```python
#If there are infinty values present in data than replacing them with null values
data.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
#checking if there are any null values
data.isna().sum()
```

# From this we can say that there were infinity values were present in the data.so afterward replaced them with null values

```python
#filling null values with mean data of that particular column
data['Family']=data['Family'].fillna(data['Family'].mean())
data['Health (Life Expectancy)']=data['Health (Life Expectancy)'].fillna(data['Health (Life Expectancy)'].mean())
data['Trust (Government Corruption)']=data['Trust (Government Corruption)'].fillna(data['Trust (Government Corruption)'].mean())
data['Generosity']=data['Generosity'].fillna(data['Generosity'].mean())
data.isna().sum()
```

```python
#checking the distribution
plt.figure(figsize=(20,25),facecolor='red')
plotnumber=1
for column in df:
    if plotnumber<=10:
        ax=plt.subplot(4,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

# By looking at the plots we can say that they are less skewed compared to the data before log transformation 

```python
# splitting data into features and labels removing country and region as i dont feel that they are more related to happiness score and removing health column because of multicolinearity with column economy
x=data.drop(['Happiness Score','Country','Region','Health (Life Expectancy)'],axis=1)
y=data['Happiness Score']
```

```python
#Importing required libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
#Initiating Decision Tree Classifier
lr=LinearRegression()
```

```python
#using range fucntion to find the best random state
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    lr.fit(x_train,y_train)
    pred=lr.predict(x_train)
    y_pred=lr.predict(x_test)
    print(f'at random state {i}, training accuracy is: {metrics.r2_score(y_train,pred)*100}')
    print(f'at random state {i}, testing accuracy is: {metrics.r2_score(y_test,y_pred)*100}')
```

```python
#since randome state  has highest testing score,so taking random state as 5
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
#Training the Model
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_train)
pred
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
#Printing the training and testing score
print('Training Score:',metrics.r2_score(y_train,pred)*100)
print('Testing Score:',metrics.r2_score(y_test,y_pred)*100)
```

# ince the training and testing is 99.13 and 99.13 this model can be overfiited. so will be using lassocv nad ridgecv method to check if the model is overfitted or not

```python
#Plotting the best fit line
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=y_pred,color='red')
plt.plot(y_test,y_test,color='blue')
plt.xlabel('Actual Happiness score',fontsize=10)
plt.ylabel('Predicted Hapiness score',fontsize=10)
plt.show()
```

```python
#imporing libraries
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
#mean absolute error
mean_absolute_error(y_test,y_pred)
```

```python
#mean squared error
mean_squared_error(y_test,y_pred)
```

```python
#Root mean squared error
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
#using lassocv and ridgecv to see of the model is overfitted or not
#importing required libraries
from sklearn.linear_model import LassoCV,Lasso,RidgeCV,Ridge
```

```python
lasscv=LassoCV(alphas=None,max_iter=100,normalize=True)
lasscv.fit(x_train,y_train)
```

```python
alpha=lasscv.alpha_
alpha
```

```python
lasso_reg=Lasso(alpha)
lasso_reg.fit(x_train,y_train)
lasso_reg.score(x_test,y_test)
```

# since the lasso reg shows the same result as of testing score so the model is not overfitted. but to confirm more precisely using ridge cv method

```python
ridgecv=RidgeCV(alphas=np.arange(0.001,0.1,0.01),normalize=True)
ridgecv.fit(x_train,y_train)
```

```python
alpha=ridgecv.alpha_
alpha
```

```python
ridge_reg=Ridge(alpha)
ridge_reg.fit(x_train,y_train)
ridge_reg.score(x_test,y_test)
```

# Ridgecv also shows the same score as of testing score so we can say that the model is not overfitted
